```python
import import_ipynb
import Linear_Regression_Model
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import r2_score
from sklearn.datasets import make_regression
```

```python
np.random.seed(0) 
# creating dataset
x, y = make_regression(n_samples=100, n_features=1, noise=10) 
y = y**2 
plt.scatter(x[:,0], y)
x.shape 
```

```python
# adding polynomial features to our dataset
poly_features = PolynomialFeatures(degree=2, include_bias=True)
x_trans = poly_features.fit_transform(x)
x_trans.shape
```

```python
theta = np.random.randn(3, 1)
y = y.reshape(y.shape[0], 1)
theta_final, cost_history, theta_history = Linear_Regression_Model.gradient_descent(x_trans, y, theta, learning_rate=0.2, iterations=10)

plt.plot(cost_history)
plt.xlabel('Iterations')
plt.ylabel('Cost')
plt.title('learning curve')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(x[:, 0], y, color='blue', label='Original data')
plt.scatter(x[:, 0],Linear_Regression_Model.model(theta_final,x_trans), color='red', label='Fitted curve')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Polynomial Regression')
plt.grid(True)
plt.legend()
plt.show()
```
