```python
np.random.seed(0) 
# creating dataset
x, y = make_regression(n_samples=100, n_features=1, noise=10) 
y = y**2 
# adding polynomial features to our dataset
poly_features = PolynomialFeatures(degree=2, include_bias=False) 
x = poly_features.fit_transform(x) 
plt.scatter(x[:,0], y) 
x.shape 
```
